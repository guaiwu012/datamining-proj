```python
# 首先，步频，提特征
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns

os.chdir(r'Z:\main\Data Mining\College Experience Study Dataset')

def map(data,MIN,MAX):
    """
    归一化映射到任意区间
    :param data: 数据
    :param MIN: 目标数据最小值
    :param MAX: 目标数据最小值
    :return:
    """
    d_min = np.max(data)    # 当前数据最大值
    d_max = np.min(data)    # 当前数据最小值
    if d_min == d_max:
        return [5]*len(data)
    return MIN +(MAX-MIN)/(d_max-d_min) * (data - d_min)
```

```python
import sys
from importlib import reload
sys.path.insert(0,r'Z:\main\Data Mining\College Experience Study Dataset')
import random
import final_dtw
reload(final_dtw)
# 首先，步频，提特征
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns

os.chdir(r'Z:\main\Data Mining\College Experience Study Dataset')

def map(data,MIN,MAX):
    d_min = np.max(data)    # 当前数据最大值
    d_max = np.min(data)    # 当前数据最小值
    if d_min == d_max:
        return [5]*len(data)
    return MIN +(MAX-MIN)/(d_max-d_min) * (data - d_min)

def post_processing2(EMA_type,sensing_type):
    time = pd.read_csv('./tried/merge.csv')
    time = time[['day','relative time']]
    time = time.drop_duplicates()
    EMA = pd.read_csv('final.EMA.'+ EMA_type +'.csv')
    sensing = pd.read_csv('final.'+ sensing_type +'.csv')
    name = list(set(sensing['uid']))
    new_sensing = pd.DataFrame(columns=[x for x in range(len(sensing.columns))])
    for x in name:
        backup = sensing[sensing['uid'] == x]
        backup[sensing_type] = map(backup[sensing_type],0,10)
        backup.columns = [x for x in range(len(sensing.columns))]
        new_sensing = pd.concat([new_sensing,backup],ignore_index=True)
    new_sensing.columns = sensing.columns
    sensing = new_sensing
    test = pd.merge(EMA,sensing,on=['uid','day'],how='outer')
    test = pd.merge(test,time,on=['day'],how='outer')
    test = test.dropna(subset=['uid'],axis=0)
    correlation = final_dtw.dtw_correlation_with_cluster2(test,'relative time',EMA=EMA_type,EMA_cluster='EMA_cluster',sensing=sensing_type,sensing_cluster='sensing_cluster')
    df_max = max(correlation.max())
    for y in correlation.columns:
        for x in correlation.index:
            if correlation.loc[x,y] == 0:
                correlation.loc[x,y] = df_max
    for y in correlation.columns:
        for x in correlation.index:
                correlation.loc[x,y] = correlation.loc[x,y] * correlation.loc[x,y]
    bg_correlation1 = final_dtw.dtw_correlation_with_background(test,'relative time',EMA=EMA_type,EMA_cluster='EMA_cluster',sensing=sensing_type,sensing_cluster='sensing_cluster')
    df_max = max(bg_correlation1.max())
    for y in bg_correlation1.columns:
        for x in bg_correlation1.index:
            if bg_correlation1.loc[x,y] == 0:
                bg_correlation1.loc[x,y] = df_max
    bg_correlation2 = final_dtw.dtw_correlation_with_background(test,'relative time',EMA=EMA_type,EMA_cluster='EMA_cluster',sensing=sensing_type,sensing_cluster='sensing_cluster')
    df_max = max(bg_correlation2.max())
    for y in bg_correlation2.columns:
        for x in bg_correlation2.index:
            if bg_correlation2.loc[x,y] == 0:
                bg_correlation2.loc[x,y] = df_max
    bg_correlation = bg_correlation2.add(bg_correlation1).div(2).add(4)
    bg_min = np.min(bg_correlation)
    bg_max = np.max(bg_correlation)
    for y in bg_correlation.columns:
        for x in bg_correlation.index:
            if bg_correlation.loc[x,y] == 0:
                bg_correlation.loc[x,y] = 5*(bg_correlation.loc[x,y]-bg_min)/(bg_max-bg_min)

    shuffle_division = correlation.divide(bg_correlation)
    '''
    for x in list(correlation.index):
        for y in list(correlation.columns):
            correlation.loc[x,y] = np.log10(correlation.loc[x,y])
    '''
    plt.figure(dpi=100,figsize = (30,10))
    plt.subplot(1,3,1)
    correlation = np.float64(correlation)
    correlation = correlation.T
    sns.heatmap(correlation,cmap="Blues_r")
    plt.xlabel(sensing_type + ' cluster')
    plt.ylabel(EMA_type + ' cluster')
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.title(EMA_type +' with ' + sensing_type + ' in real cluster')

    plt.subplot(1,3,2)
    bg_correlation = np.float64(bg_correlation)
    bg_correlation = bg_correlation.T
    sns.heatmap(bg_correlation,cmap="Blues_r")
    plt.xlabel(sensing_type + ' cluster')
    plt.ylabel(EMA_type + ' cluster')
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.title(EMA_type +' with ' + sensing_type + ' in shuffle')

    plt.subplot(1,3,3)
    shuffle_division = np.float64(shuffle_division)
    shuffle_division = shuffle_division.T
    sns.heatmap(shuffle_division,cmap="Blues_r")
    plt.xlabel(sensing_type + ' cluster')
    plt.ylabel(EMA_type + ' cluster')
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.title(EMA_type +' with ' + sensing_type + ' in division')
    plt.tight_layout
    plt.savefig(EMA_type + '-' + sensing_type + '.png')
```
