```python
import os
import pandas as pd

dataset_directory = 'College Experience Study Dataset'

for filename in os.listdir(dataset_directory):
    if filename.endswith(".csv"):
        file_path = os.path.join(dataset_directory, filename)
        df = pd.read_csv(file_path)
        
        col_4_values = df.iloc[:, 3].dropna()  # 删除缺失值
        
        # 计算最小值和最大值
        min_value = col_4_values.min()
        max_value = col_4_values.max()
        
        # 输出结果
        print(f"For file {filename}:")
        print(f"Minimum value in column 4: {min_value}")
        print(f"Maximum value in column 4: {max_value}")
        print()
```

```python
import os
import pandas as pd
import numpy as np

dataset_directory = 'College Experience Study Dataset'
output_directory = 'College Experience Study Dataset Shuffled'

if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# 遍历目录下的所有文件
for filename in os.listdir(dataset_directory):
    if filename.endswith(".csv"):
        file_path = os.path.join(dataset_directory, filename)
        df = pd.read_csv(file_path)
        
        col_4_name = df.columns[3]
        
        grouped = df.groupby('uid')[col_4_name]
        min_max_values = grouped.agg([np.min, np.max]) 
        min_max_values['new_col_4'] = np.random.randint(min_max_values['amin'], min_max_values['amax'] + 1, size=len(min_max_values))
        
        df = pd.merge(df, min_max_values[['new_col_4']], left_on='uid', right_index=True)
        df[col_4_name] = df['new_col_4']
        df = df.drop(columns=['new_col_4'])
        
        output_file_path = os.path.join(output_directory, filename)
        df.to_csv(output_file_path, index=False)
        
        print(f"Random values generated and replaced, and the file is saved to: {output_file_path}")
```
